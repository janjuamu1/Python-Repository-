#### List Comprehension
List comprehension is a concise readable way to create a list in python. It allows you to generate a new list by applying an expression to each element in an existing table, like a tuple, list or a range or objects. 
#### Syntax 
'[expression for item in iterable if condition)
* `expression`: this is the operation you perform on each item. The output from this expression is what will be included in the new list. 
* 'item' : This is a variable that takes the value of each element from the 'iterable' as the list comprehension loops through it
* 'iterable' : Any python object that can be looped over (lists, tuples, dictionaries, sets, ranges, etc) 
* 'condition': Optional. Items for which the condition evaluates to 'True' are processed by the expression 

#### Example 1: Create a list of squares 

```python
squareList = [x**2 for x in range(10)]
squareList 
```

```python
for x in range(10):
    print(x**2, end=" ")
```

#### Example 2: Filter even numbers and square them

```python
for x in range(10):
    if x % 2 == 0:
        print(x**2)
```

```python
squareEvenList = [x**2 for x in range(10) if x % 2 == 0]
squareEvenList
```

#### Example 3: Convert temperatures from Celsius to Fahrenheit 
* Formula is: '(9/5)*CelsiusTemp + 32

```python
celsius = [0, 10, 20, 35, 38]
fahrenheit = [((9/5)*celsiusTemp + 32) for celsiusTemp in celsius]
fahrenheit
```
